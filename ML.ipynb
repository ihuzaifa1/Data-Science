```python
from sklearn.datasets import load_breast_cancer
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = load_breast_cancer()
```

```python
dataset = pd.DataFrame(df.data)
```

```python
dataset
```

```python
dataset.columns = df.feature_names
```

```python
dataset.head()
```

```python
X = dataset
y = df.target
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=42)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
```

```python
X_train= scalar.fit_transform(X_train)
```

```python
X_test = scalar.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
```

```python
regression = LinearRegression()
regression.fit(X_train,y_train)
```

```python
mse = cross_val_score(regression,X_train,y_train,scoring='neg_mean_squared_error',cv=10)
```

```python
np.mean(mse)
```

```python
reg_pred = regression.predict(X_test)
```

```python
reg_pred
```

```python
import seaborn as sns
sns.displot(reg_pred-y_test,kind='kde')
```

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(reg_pred,y_test)
```

```python
score
```

```python
#Ridge & lasso
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
ridgeregression = Ridge()
```

```python
ridgeregression
```

```python
parameters = {'alpha':[1,2,3,5,10,20,30,40,50,60,100]}
ridgecv = GridSearchCV(ridgeregression,parameters,scoring='neg_mean_squared_error',cv=5)
ridgecv.fit(X_train,y_train)
```

```python
print(ridgecv.best_params_)
```

```python
print(ridgecv.best_score_)
```

```python
ridge_predict = ridgecv.predict(X_test)
```

```python
import seaborn as sns
sns.displot(ridge_predict-y_test,kind='kde')
```

```python
score_ridge = r2_score(ridge_predict,y_test)
```

```python
score_ridge
```
